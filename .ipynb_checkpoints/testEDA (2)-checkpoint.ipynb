```python
##The provided code snippet performs oversampling using SMOTE (Synthetic Minority Over-sampling Technique) to address class imbalance in the dataset.
```

```python
from imblearn.over_sampling import SMOTE
selected_variables = ['orX','orY','orZ','rX','rY','rZ','accX','accY','accZ','gX','gY','gZ','mX','mY','mZ']
x = df[selected_variables]
y= df['activity']
smote = SMOTE(random_state=42)
x,y = smote.fit_resample(x,y)
```

```python
#displaying the shape of x
```

```python
x.shape
```

```python
#displyaying the shape of y
```

```python
y.shape
```

```python
##here we join x and y 
```

```python
data = pd.concat([x, y], axis=1)
```

```python
#here we  encode categorical values in the 'activity' column of a DataFrame data as numeric labels. 
```

```python
data['activity'] = pd.factorize(data['activity'])[0]
```

```python
##checking for the number of categories in the activity column
```

```python
data['activity'].unique()
```

```python
# Variable dependencies
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Variable Dependencies")
plt.show()
```
